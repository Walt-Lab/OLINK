```python
# clean up the raw olink data file

assay_list = pd.read_excel(ASSAY_LIST_FILE)
plate_layout = pd.read_excel(PLATE_LAYOUT_FILE)
raw_data = pd.read_csv(RAW_DATA_FILE, sep=";")

raw_data = raw_data.drop(UNIMPORTANT_COLUMNS, axis=1)

layout_dict = dict(zip(plate_layout.Location, plate_layout.Sample))
layout_dict2 = dict(zip(plate_layout.Location, plate_layout.Additional_info))

raw_data["Sample Description"] = raw_data["WellID"].map(layout_dict)
raw_data["Additional Information"] = raw_data["WellID"].map(layout_dict2)


def make_linear(x):
    return 2**x


raw_data["Linear NPX"] = raw_data["NPX"].map(make_linear)
```

```python
# compare the internal cargo in plasma and CSF

internal_data = raw_data[raw_data["Sample Description"].str.contains("EV")]

plas_data = internal_data[
    raw_data["Additional Information"] == "EV internal cargo in plasma"
]
csf_data = internal_data[
    raw_data["Additional Information"] == "EV internal cargo in CSF"
]

plas_medians = plas_data.groupby("Assay")["NPX"].median()
csf_medians = csf_data.groupby("Assay")["NPX"].median()

all_medians = pd.concat([plas_medians, csf_medians], axis=1)
all_medians.columns = ["Plasma Mean NPX", "CSF Mean NPX"]

plasma_greater_npx = all_medians[
    all_medians["Plasma Mean NPX"] > all_medians["CSF Mean NPX"]
]
csf_greater_npx = all_medians[
    all_medians["CSF Mean NPX"] > all_medians["Plasma Mean NPX"]
]

plasma_greater_npx.to_excel(PLASMA_GREATER_NPX_PATH)
csf_greater_npx.to_excel(CSF_GREATER_NPX_PATH)
```

```python
# Parse apart the data from the early fractions vs. the late fractions
ev_fractions_subset = raw_data[
    (raw_data["Sample Description"].str.contains("Fract 7"))
    | (raw_data["Sample Description"].str.contains("Fract 8"))
    | (raw_data["Sample Description"].str.contains("Fract 9"))
    | (raw_data["Sample Description"].str.contains("Fract 10"))
]

late_fractions_subset = raw_data[
    (raw_data["Sample Description"].str.contains("Fract 11"))
    | (raw_data["Sample Description"].str.contains("Fract 12"))
    | (raw_data["Sample Description"].str.contains("Fract 13"))
    | (raw_data["Sample Description"].str.contains("Fract 14"))
    | (raw_data["Sample Description"].str.contains("Fract 6"))
]

ev_fractions_sums = ev_fractions_subset.groupby("Assay")["Linear NPX"].sum()
late_fractions_sums = late_fractions_subset.groupby("Assay")["Linear NPX"].sum()

all_sums = pd.concat([ev_fractions_sums, late_fractions_sums], axis=1)
all_sums.columns = ["EV Fractions", "Late Fractions"]
all_sums = all_sums.rename_axis("Assay").reset_index()
```

```python
# t-test to determine statistical significance of differences in early v. late fractions

assay_names = []
p_values = []


for assay in raw_data["Assay"].unique():
    subset = raw_data[raw_data["Assay"] == assay]
    ev_fractions = subset[
        subset["Sample Description"].str.contains("Fract 7|Fract 8|Fract 9|Fract 10")
    ]["Linear NPX"]
    late_fractions = subset[
        subset["Sample Description"].str.contains("Fract 11|Fract 12|Fract 13|Fract 14")
    ]["Linear NPX"]

    t_stat, p_value = stats.ttest_rel(ev_fractions, late_fractions)

    assay_names.append(assay)
    p_values.append(p_value)

result_data = {"Assay": assay_names, "p-value": p_values}
result_df = pd.DataFrame(result_data)

sig_dif = result_df[result_df["p-value"] < SIGNIFICANCE_VALUE]
# sig_dif = pd.merge(sig_dif, all_sums, on="Assay", how="left")
# filtered_sig_dif = sig_dif[sig_dif["EV Fractions"] > sig_dif["Late Fractions"]]
```
